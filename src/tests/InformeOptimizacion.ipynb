```python
import matplotlib.pyplot as plt
import numpy as np
from random import random
```

```python
def parseTimes(times: str) -> list[float]:
    while '  ' in times:
        times = times.replace('  ', ' ')
    return [float(num) for num in times.split()]
```

```python
def printChart(optimizado: list[float], noOptimizado: list[float], title: str) -> None:
    labels = [str(i+1) for i in range(len(optimizado))]

    x = np.arange(len(labels))
    width = 0.35

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, optimizado, width, label="Sin poda", color="#388697")
    rects2 = ax.bar(x + width/2, noOptimizado, width, label="Con poda", color="#08415C")

    ax.set_ylabel('Tiempo de ejecucion (ms)')
    ax.set_xlabel('Cantidad de candidatas')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()

    ax.bar_label(rects1, padding=3)
    ax.bar_label(rects2, padding=3)

    fig.tight_layout()

    plt.yscale('log')
    plt.show()
```

```python
def relacionPromedio(data1: list[float], data2: list[float]) -> float:
    diferencias = [(b/a)*100 for a,b in zip(data1, data2)]
    return sum(diferencias) / len(diferencias)
```

# Mejor opci√≥n al principio

```python
primeraNoPodaStr = "0.594100    0.883200    1.240600    0.781500    2.108500    2.708600    3.878400    4.072200    4.510100    7.784600"
primeraPodaStr = "0.038500    0.009300    0.010300    0.008700    0.013500    0.020200    0.034100    0.023200    0.016400    0.033300"

primeraNoPoda = parseTimes(primeraNoPodaStr)
primeraPoda = parseTimes(primeraPodaStr)
diferencia = relacionPromedio(primeraNoPoda, primeraPoda)

printChart(primeraNoPoda, primeraPoda, 'Mejor opci√≥n al principio.')
print(f'En promedio la versi√≥n con poda del algoritmo se ejecuta en el {diferencia:.2f}% del tiempo que consume la versi√≥n sin poda.')
```
